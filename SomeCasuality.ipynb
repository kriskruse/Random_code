```python
import numpy as np
import pandas as pd
from itertools import combinations
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
df = pd.read_csv("data_483.csv")

lst = ['A','B','C','D']
combi = combinations(lst,2)

for item1, item2 in combi:
    print("------" + item1, item2 + "--------")
    grangercausalitytests(df[[item1, item2]], maxlag=10)
    print("")
    print("reversed")
    grangercausalitytests(df[[item2, item1]], maxlag=10)
```
